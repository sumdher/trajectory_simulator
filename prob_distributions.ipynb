```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

a, b = 1.9, 4.0
lo, hi = 11, 180

x_unit = np.linspace(0, 1, 500)
x_scaled = lo + x_unit * (hi - lo)
pdf = beta.pdf(x_unit, a, b) / (hi - lo)   # divide to keep area = 1

mode = lo + (a - 1) / (a + b - 2) * (hi - lo)

plt.plot(x_scaled, pdf)
plt.xlabel("Dwell time (s)")
plt.ylabel("Density")
plt.title(f"Beta({a}, {b}) scaled to [{lo}, {hi}]")
plt.axvline(lo + (a-1)/(a+b-2) * (hi-lo), color='red', linestyle='--', label=f'mode: {mode}')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

phases = {
    "WALK": 0.75,
    "APPROACH": 0.30,
    "STOP": 0.66,
}

sigma_max = max(phases.values())
x = np.linspace(-3 * sigma_max, 3 * sigma_max, 500)

for phase, sigma in phases.items():
    pdf = norm.pdf(x, loc=0, scale=sigma)
    plt.plot(x, pdf, label=f"{phase}  std = {sigma}")

plt.xlabel("speed (m/s)")
plt.ylabel("Density")
plt.title("Normal(0, std_dev)")
plt.legend()
plt.show()

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

phases = {
    "WALK": 0.80,
    "APPROACH": 0.50,
    "STOP": 0.75,
}

sigma_max = max(phases.values())
x = np.linspace(-3 * sigma_max, 3 * sigma_max, 500)

for phase, sigma in phases.items():
    pdf = norm.pdf(x, loc=0, scale=sigma)
    plt.plot(x, pdf, label=f"{phase}  std = {sigma}")

plt.xlabel("del_theta")
plt.ylabel("Density")
plt.title("Normal(0, std_dev)")
plt.legend()
plt.show()
```
